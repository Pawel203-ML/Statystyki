```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
def gauss(x, sigma = 1, mu = 3):
    return stats.norm.pdf(x, mu, sigma)
```

```python
x10000 = np.linspace(-10, 10, 10000)
x100 = np.linspace(-10,10, 100)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x10000, gauss(x10000), label = '10000 wartosci', alpha = 0.3)
plt.plot(x100, gauss(x100), label = '100 wartosci', alpha = 0.3)
plt.legend()
plt.title('Rozklad nomalny dla sredniej rownej 3 oraz odchyleniu standardowym rownym 1')
plt.show()
```

```python
#gauss
gauss_100 = gauss(x100)
gauss_10000  = gauss(x10000)
```

# gauss_100

```python
plt.figure(figsize = (15,10))
plt.hist(gauss_100, bins = 30)
plt.axvline(x = np.mean(gauss_100), color = 'red', label = 'Mean', alpha = 0.7)
plt.axvline(x = np.median(gauss_100), color = 'black', label = 'Median',alpha = 0.7)
plt.axvline(x = stats.mode(gauss_100)[0], color = 'green', label = 'Mode', alpha = 0.7)
plt.axvline(x = min(gauss_100), color = 'yellow', label = 'q0', alpha = 0.7)
plt.axvline(x = np.quantile(gauss_100, 0.25), color = 'yellow', label = 'q1', alpha = 0.7)
plt.axvline(x = np.quantile(gauss_100, 0.75), color = 'yellow', label = 'q3', alpha = 0.7)
plt.axvline(x = max(gauss_100), color = 'yellow', label = 'q4', alpha = 0.7)
plt.legend(loc = 'upper center')
plt.title('statystyki dla zbioru gauss_100')
plt.show()
```

```python
a = stats.mode(gauss_100)[0] # green
b = np.median(gauss_100) # black
print(a < b)
```

Wnioski: Asymetria rozkladu jest dodatnia

```python
# Zakres
range_gause_100 = max(gauss_100) - min(gauss_100)
# IQR
iqr = np.quantile(gauss_100, 0.75) - np.quantile(gauss_100, 0.25)
# Wariancja
variance_gauss_100 = np.var(gauss_100, ddof = 1)
print(f'Zakres: {range_gause_100} \n IQR: {iqr} \n Wariancja: {variance_gauss_100}')
```

## Normalizacja

```python
scaler_MinMax = MinMaxScaler()
normalized_gauss_100 = scaler_MinMax.fit_transform(gauss_100.reshape(-1,1))
plt.hist(normalized_gauss_100, bins = 20)
plt.show()
```
